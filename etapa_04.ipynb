# Etapa 04) Comparações por País

```python
import pandas as pd

# Importando a base
df = pd.read_csv('salario_profissionais_dados.csv')
```

```python
# Quais são os 10 países com as maiores médias salariais?
# Resultado: Este comando listará os países onde os profissionais de dados recebem as maiores remunerações anuais. Geralmente, países com moedas fortes ou polos tecnológicos (como EUA, Suíça ou Israel) aparecem no topo.

# Agrupando por país e calculando a média salarial
# Usamos sort_values(ascending=False) para colocar os maiores salários no topo
media_por_pais = df.groupby('country')['salary_in_usd'].mean().sort_values(ascending=False)

# Exibindo os 10 primeiros
print(media_por_pais.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Criando um gráfico de barras para os 10 países com maiores médias

plt.figure(figsize=(12, 6))
top_10_paises = media_por_pais.head(10)

sns.barplot(x=top_10_paises.values, y=top_10_paises.index, palette='Blues_r')

plt.title('Top 10 Países com Maiores Médias Salariais (USD)')
plt.xlabel('Média Salarial Anual (USD)')
plt.ylabel('País')
plt.show() 
```

```python
# Filtrando apenas o Brasil para comparar com o topo do ranking

media_brasil = df[df['country'] == 'Brazil']['salary_in_usd'].mean()
print(f"Média Salarial no Brasil: US$ {media_brasil:.2f}")
```
