```python
import pandas as pd

# Importando a base
df = pd.read_csv('salario_profissionais_dados.csv')
```

```python
# Quais são as medidas de tendência central (Média e Mediana)?
# Resultado: A Média indica o valor "balanceado" de todos os salários, enquanto a Mediana mostra o valor exatamente no meio da amostra. Se a média for muito maior que a mediana, temos salários muito altos puxando o resultado.

# Calculando a média e a mediana salarial
media = df['salary_in_usd'].mean()
mediana = df['salary_in_usd'].median()

print(f"Média Salarial: US$ {media:.2f}")
print(f"Mediana Salarial: US$ {mediana:.2f}")
```

```python
# Quais são as medidas de dispersão (Desvio Padrão, Mínimo e Máximo)?
# Resultado: O Desvio Padrão revela o quanto os salários variam. Um desvio alto indica que os salários estão muito espalhados (muita desigualdade ou cargos muito diferentes).

# Calculando desvio padrão, mínimo e máximo
desvio = df['salary_in_usd'].std()
minimo = df['salary_in_usd'].min()
maximo = df['salary_in_usd'].max()

print(f"Desvio Padrão: US$ {desvio:.2f}")
print(f"Salário Mínimo: US$ {minimo:.2f}")
print(f"Salário Máximo: US$ {maximo:.2f}")
```

```python
# Como é a distribuição geral dos salários (Histograma)?
# Resultado: O histograma permite visualizar se a maioria dos salários está concentrada em valores baixos, médios ou altos.

import seaborn as sns
import matplotlib.pyplot as plt

# Criando um histograma com a linha de densidade (KDE)
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], kde=True, color='teal')
plt.title('Distribuição de Salários em USD')
plt.xlabel('Salário Anual (USD)')
plt.ylabel('Frequência')
plt.show()
```

```python
# Como os salários se comparam por nível de experiência (Boxplot)?
# Resultado: O Boxplot é excelente para ver a variação dentro de cada nível (EN=Junior, MI=Pleno, SE=Senior, EX=Executivo). Os pontos fora das "hastes" são os outliers (salários fora do comum).

# Criando um boxplot para comparar os níveis de experiência
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, 
            order=['EN', 'MI', 'SE', 'EX'], palette='Set3')
plt.title('Comparativo Salarial por Nível de Experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário em USD')
plt.show()
```

```python
# Qual a média salarial por tamanho de empresa?
# Resultado: Mostra se empresas Grandes (L), Médias (M) ou Pequenas (S) pagam melhor em média.

# Usando groupby para comparar as médias por tamanho da empresa
media_por_tamanho = df.groupby('company_size')['salary_in_usd'].mean().sort_values()
print(media_por_tamanho)
```
