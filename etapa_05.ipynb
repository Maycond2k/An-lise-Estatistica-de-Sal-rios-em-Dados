# Etapa 05) Correlações e Tendências

```python
import pandas as pd

# Importando a base
df = pd.read_csv('salario_profissionais_dados.csv')
```

```python
# Existe alguma tendência de aumento salarial com os anos?
# Resultado: Este gráfico mostra se, de 2020 até 2023, o mercado de dados valorizou e os salários médios subiram globalmente.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Agrupando a média salarial por ano (work_year)
tendencia_anual = df.groupby('work_year')['salary_in_usd'].mean()

# 2. Criando um gráfico de linha para ver a evolução
plt.figure(figsize=(8, 5))
sns.lineplot(x=tendencia_anual.index, y=tendencia_anual.values, marker='o', color='green')
plt.title('Tendência Salarial ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Média Salarial (USD)')
plt.xticks(tendencia_anual.index)
plt.show()
```

```python
# Quanto maior o tempo de experiência, maior o salário? (Matriz de Correlação)

# Resultado: O coeficiente de correlação varia de -1 a 1.
# Se a relação entre years_of_experience e salary_in_usd estiver próxima de 1, significa que quanto mais experiência, maior o salário (correlação positiva forte).
# Se estiver próxima de 0, o tempo de experiência não garante necessariamente um aumento salarial na base.

# 1. Selecionando apenas as colunas numéricas para a correlação
colunas_numericas = df[['salary_in_usd', 'work_year', 'years_of_experience']]
matriz_correlacao = colunas_numericas.corr()

# 2. Exibindo a matriz
print(matriz_correlacao)

# 3. Criando um Heatmap (Mapa de Calor) para facilitar a leitura
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação entre Variáveis Numéricas')
plt.show()
```

```python
# Visualizando a relação direta (Gráfico de Dispersão)
# Resultado: A linha vermelha subindo confirma visualmente a tendência de que a senioridade é recompensada financeiramente.

# Criando um gráfico de dispersão com linha de tendência
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='years_of_experience', y='salary_in_usd', line_kws={"color": "red"})
plt.title('Relação: Anos de Experiência vs Salário')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário (USD)')
plt.show()
```
