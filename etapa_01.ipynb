# Etapa 01) Importação e Exploração Inicial da Base

```python
# Quais colunas estão disponíveis?
# Resultado: As colunas são work_year, country, region, experience_level, job_title, salary_in_usd, employee_residence, company_location, company_size e years_of_experience.

import pandas as pd

# Importando a base
df = pd.read_csv('salario_profissionais_dados.csv')

# Exibindo os nomes das colunas
print(df.columns) 
```

```python
# Quantas linhas existem?
# Resultado: O dataset possui 3.755 linhas.

# Verificando a quantidade de linhas
print(len(df))
# Ou usando shape (Linhas, Colunas)
print(df.shape[0])
```

```python
# Existem dados ausentes?
# Resultado: Não existem dados ausentes. Todas as colunas apresentam o valor 0 para dados nulos.

# Verificando a soma de valores nulos por coluna
print(df.isnull().sum())
```

```python
# Quais são os tipos de variáveis?
# Resultado: > * int64 (Numéricas): work_year, salary_in_usd, years_of_experience.
# object (Texto/Categóricas): country, region, experience_level, job_title, employee_residence, company_location, company_size.

# Verificando os tipos de dados de cada coluna
print(df.dtypes)
```
