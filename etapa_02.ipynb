# Etapa 02) Frequência e Distribuição das Categorias

```python
import pandas as pd

# Importando a base
df = pd.read_csv('salario_profissionais_dados.csv')
```

```python
# Quais são os 10 cargos mais comuns?
# Resultado: Os cargos que mais aparecem na base são Data Engineer, Data Scientist e Data Analyst. Juntos, esses três representam a grande maioria dos profissionais no dataset.

# Contagem de profissionais por cargo (Top 10)
cargos_frequentes = df['job_title'].value_counts().head(10)
print(cargos_frequentes)

# Para ver em porcentagem (proporção relativa)
cargos_porcentagem = df['job_title'].value_counts(normalize=True).head(10) * 100
print(cargos_porcentagem)
```

```python
# Qual a distribuição por nível de experiência?
# Resultado: A base é predominantemente composta pelo nível SE (Senior), seguido por MI (Mid-level/Intermediário). Os níveis EN (Entry-level) e EX (Executive) possuem menor volume de dados.

# Contagem por nível de experiência
experiencia_freq = df['experience_level'].value_counts()
print(experiencia_freq)

# Em porcentagem
experiencia_pct = df['experience_level'].value_counts(normalize=True) * 100
print(experiencia_pct)
```

```python
# Qual o perfil das empresas pelo tamanho?
# Resultado: A maioria esmagadora das oportunidades está concentrada em empresas de tamanho M (Medium), seguidas pelas empresas L (Large). Empresas S (Small) são a minoria no dataset.

# Contagem por tamanho da empresa (S: Small, M: Medium, L: Large)
tamanho_empresa_freq = df['company_size'].value_counts()
print(tamanho_empresa_freq)

# Em porcentagem
tamanho_empresa_pct = df['company_size'].value_counts(normalize=True) * 100
print(tamanho_empresa_pct)
```
